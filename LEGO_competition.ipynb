```python
#import file sets.csv

import pandas as pd
import altair as alt
# import numpy as np

setsdf = pd.read_csv('sets.csv')
themesdf = pd.read_csv('themes_grouped.csv')
```

```python
# LEGOdf = pd.concat(
#     [setsdf, themesdf],
#     keys="theme_id",
# )

LEGOdf = pd.merge(
    setsdf,
    themesdf,
    how="left",
    on="theme_id",
    sort=True
)
```

```python
LEGOdf
```

```python
LEGO_group_df = LEGOdf.groupby(["year", "theme-group-name"])['num_parts'].sum().reset_index()
LEGO_group_df = LEGOdf.groupby(["year", "theme-group-name"]).size().to_frame('size').reset_index()
```

```python
LEGO_group_df['year']
```

```python
alt.Chart(LEGO_group_df).mark_bar().encode(
#     x='num_parts',
    x='size',
    y=alt.X('theme-group-name').sort('-x')
)
```

```python
# topten = ["Star Wars", "Town", "Technic", "Creator", "Education and Dacta", 
#           "Universal Building Set", "Superheroes", "Sculptures", "Space", "Train"]

topten = ["Town", "Seasonal", "Star Wars", "Technic", "Service Packs", 
          "Universal Building Set", "Collectible Minifigures", "Gear", "Creator", 
          "Bionicle", "Classic", "Educational and Dacta", "Space", "Train", "Racers", 
          "Castle"]

topten = ["Town", "Seasonal", "Star Wars", "Technic", "Service Packs", 
          "Universal Building Set", "Collectible Minifigures", "Gear"]

LEGO_10_df = LEGO_group_df[LEGO_group_df["theme-group-name"].isin(topten)]
```

```python
LEGO_10_df
```

```python
LEGO_10_df.groupby("theme-group-name")["year"].mean()

LEGO_10_order = [
    "Collectible Minifigures",    #2012
    "Star Wars",                  #2008
    "Gear",                       #2004
    "Seasonal",                   #2001
    "Technic",                    #1997
    "Town",                       #1997
    "Service Packs",              #1990
    "Universal Building Set",     #1986
                ]

LEGO_10_order.reverse()
```

```python
LEGO_10_df.groupby("theme-group-name").sum()
# Each LEGO is 20 or 25?
```

```python
base = alt.Chart(LEGO_10_df).mark_bar().encode(
    x=alt.X('year_bins:O'),
    y='sum_sets:Q'
).transform_bin(
    'year_bins', 'year', bin=alt.Bin(maxbins=16)
).transform_aggregate(
    sum_sets='sum(size)',
    groupby=["year_bins", "theme-group-name"]
).properties(
    width=800,
    height=100
)

bars = base.mark_bar().encode(
    y = 'sum_sets:Q',
    color='theme-group-name'
)

text = base.mark_text(dy = -5).encode(
    text='sum_sets:Q',
    y='sum_sets:Q'
)

LEGO_chart = bars + text

LEGO_chart = LEGO_chart.facet(
    row=alt.Row('theme-group-name', sort=LEGO_10_order)
)

LEGO_chart
```
